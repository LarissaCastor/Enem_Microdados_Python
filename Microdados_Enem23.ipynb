#**Questão 2:**
a) Crie uma tabela com as médias das notas (NU_NOTA_CN, NU_NOTA_CH, NU_NOTA_LC, NU_NOTA_MT) por estado (SG_UF_PROVA).


```python
#Filtrando colunas solicitadas
colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT']
```

```python
#Agrupamento das médias das notas por estado
tabela_medias = df.groupby('SG_UF_PROVA')[colunas_notas].mean()
tabela_medias.head(2)
```

#**Questão 2:**
 b) Identifique quais estados tiveram menor e maior média em cada área.

```python
#Selecionando menores médias e estados
menor_estado = tabela_medias.idxmin()
menor_valor = tabela_medias.min()
#Selecionando maiores médias e estados
maior_estado = tabela_medias.idxmax()
maior_valor = tabela_medias.max()
```

```python
#Apresentando menores e maiores médias por estado em dataframe
for coluna in colunas_notas:                                         #loop para cada área
    menor_estado = tabela_medias[coluna].idxmin()                    #retorna estado com menor média
    menor_valor = tabela_medias[coluna].min()                        #retorna menor média

    maior_estado = tabela_medias[coluna].idxmax()                    #retorna estado com maior média
    maior_valor = tabela_medias[coluna].max()                        #retorna maior média

    print(f"{coluna}:")                                              #imprime a área
    print(f"  Menor média → {menor_estado} ({menor_valor:.2f})")     #imprime menor média e estado 'float' 2 casas decimais
    print(f"  Maior média → {maior_estado} ({maior_valor:.2f})\n")   #imprime maior média e estado 'float' 2 casas decimais e quebra de linha
```

**Explicação** : É possível verificar que o estado do Maranhão apresentou menores médias em todas as áreas e o estado do Rio Grande do Norte apresentou maiores médias em todas as áreas. Uma apresentação visual e comparativa entre as médias dos estados é disponibilizada a seguir.

```python
#Gráfico de barras para visualizar a tabela_medias
tabela_medias.plot(kind='bar', figsize=(8, 4))                     #Escolhendo tipo de gráfico, tamanho
plt.xlabel('Estados')                                              #Nomeando eixo x
plt.ylabel('Média das Notas')                                      #Nomeando eixo y

plt.xticks(rotation=360)                                           #Rotacionando os rótulos do eixo X para evitar sobreposição
plt.legend(title='Áreas de Conhecimento', bbox_to_anchor=(1, 1))   #Adicionando uma legenda
plt.tight_layout()                                                 #Ajustando o layout para evitar que os rótulos sejam cortados
plt.show()                                                         #Chamando a imagem
```

#**Questão 4** : Visualização de dados
b) Crie um histograma das notas de NU_NOTA_MT para uma UF específica.



*  Para realização do histograma solicitado, como o dataset é muito grande, os dados serão agrupados por classes.
*  Foi escolhida a regra de Sturges para classificação devido ao tamanho do dataset.



```python
notas_mat = df['NU_NOTA_MT']

#Definir número de classes pela regra de Sturges
n = len(notas_mat)                                  #número de observações
k_sturges = int(1 + 3.322 * np.log10(n))            #Regra de Sturges

#Definindo o número de classes baseado na regra de Sturges
k = k_sturges

#Construir intervalos de classe
amplitude = notas_mat.max() - notas_mat.min()       #amplitude total dos dados
h = amplitude / k                                   #largura de cada classe

#Criar os intervalos de classe
bins = np.arange(notas_mat.min(), notas_mat.max() + h, h)

#Criar tabela de frequência usando pd.cut para categorizar as notas em classes
freq = pd.cut(notas_mat, bins=bins, right=False).value_counts().sort_index()

# Criar rótulos A, B, C...
rotulos = [chr(65+i) for i in range(len(freq))]     #'len(freq)'- quantidade de classes; '65+i': começa em 65 (código Unicode da letra 'A') e soma i

#Montar DataFrame da distribuição de frequência
tabela_freq = pd.DataFrame({
    "Rótulo": rotulos,
    "Classe": freq.index.astype(str),               #intervalos
    "Frequência": freq.values})                     #frequências absolutas

tabela_freq
```

**Explicação** : Observa-se que a definição de 19 classes abrangendo todo intervalo de notas [min , max) e já é possível verificar a frequência em cada classe estabelecida. A classe denominada como I representando o intervalo [383.44, 431.37) apresentou maior frequência (168286). Foi percebido também que nenhum participante obteve nota entre [47.93 , 287.58) condizente com os intervalos das classe B, C, D, E e F.

```python
#Criando HISTOGRAMA com caixa de legenda
fig, ax = plt.subplots(figsize=(12, 8))

# Desenha as barras: uma por classe (como um histograma categórico)
ax.bar(tabela_freq["Rótulo"], tabela_freq["Frequência"], edgecolor="black", alpha=0.5)

# Títulos e rótulos dos eixos
ax.set_title("Distribuição das Notas de Matemática — ENEM 2023 (Nordeste)", pad=16)
ax.set_xlabel("Classes")
ax.set_ylabel("Frequência")

#Criando eixo secundário
#Criando eixo X superior com notas mapeadas para posições das barras
STEP = 100                                                       #range desejado
ini = int(np.floor(bins[0] / STEP) * STEP)                       #início múltiplo de 100
fim = int(np.ceil(bins[-1] / STEP) * STEP)                       #fim múltiplo de 100
ticks_notas = np.arange(ini, fim + 1, STEP)                      #range de 100 em 100
pos_ticks = np.interp(ticks_notas, bins, np.arange(len(bins)))   # mapa nota -> posição contínua
mask = (pos_ticks >= 0) & (pos_ticks < len(freq))                #mantém range dentro do gráfico filtrando os vetores

ax2 = ax.twiny()                                                 #eixo X superior
ax2.set_xlim(ax.get_xlim())                                      #alinha limites com o eixo de baixo
ax2.set_xticks(pos_ticks[mask])                                  #posições dos ticks
ax2.set_xticklabels([str(int(v)) for v in ticks_notas[mask]])    #rótulos (números)
ax2.set_xlabel("Notas")                                          #rótulo do eixo X (topo)
ax2.tick_params(axis='x', pad=6)                                 #espaço entre eixos

#Criando o texto da legenda (DENTRO DO GRÁFICO)
tabela_freq["Classe"] = (tabela_freq["Classe"].astype(str).str.replace("(", "[", regex=False).str.replace("]", ")", regex=False))     #mantém o padrão da escrita da frequência [  ,  )
linhas_legenda = [f"{r}: {c}" for r, c in zip(tabela_freq["Rótulo"], tabela_freq["Classe"])]     #'f-string' para montar linhas com as classes onde pra cada interação, 'r' (rótulo) e 'c'(intervalo)

#Unindo em um bloco de texto (uma linha por classe)
texto_legenda = "\n".join(linhas_legenda)

#Adicionando a "caixa de legenda" ao lado direito, dentro da área do gráfico
ax.text(
    1.02, 0.5,                      # posição (x, y) em coordenadas relativas ao eixo
    texto_legenda,                  # conteúdo do texto
    transform=ax.transAxes,         # usa o sistema de coordenadas do eixo
    va="center", ha="left",         # alinhamentos vertical e horizontal
    fontsize=9,                     # tamanho da fonte
    bbox=dict(                      # estilo da caixa (legenda)
        boxstyle="round",           # cantos arredondados
        facecolor="white",          # fundo branco
        edgecolor="black",          # borda preta
        alpha= 0.9                  # leve transparência.
    )
)

plt.show()
```

**Explicação** : A figura a cima ratifica as informações anteriores apresentadas na tabela. Foi decidido realizar a apresentação de dois eixos 'x' para facilitar interpretações visuais. Dessa forma, é possível tanto visualizar classe a classe como já visualizar de forma mais simples as regiões mais frequentes de notas.

**OBS:** Como sugestão adicional, foi realizado um ranqueamento referente às médias de Matemática entre os estados avaliados.

```python
#Ordenar a tabela de forma decrescente para cada área
tabela_medias_reset = tabela_medias.reset_index()
tabela_medias_reset = tabela_medias_reset.rename(columns={'SG_UF_PROVA': 'Estados','NU_NOTA_MT': 'Média_MT'})

#criando um ranking para as médias de Matemática
rank_mt = tabela_medias_reset[['Estados', 'Média_MT']].sort_values(by='Média_MT', ascending=False).round(2).reset_index(drop=True)
rank_mt['Posição'] = rank_mt.index + 1          #adiciona posição no ranking
rank_mt
```

```python
#Apresentação visual do ranking
plt.figure(figsize=(10,6))
plt.bar(rank_mt['Estados'], rank_mt['Média_MT'])

plt.title('Ranking das Médias de Matemática por Estado', fontsize=14)
plt.xlabel('Estados')
plt.ylabel('Média Matemática')
plt.show()
```

**Explicação** : O ranking mostra que o Estado de Rio Grande do Norte apresntou maior média para a área de Matemática e Tecnologias, seguindo do estado do Ceará com uma sugunda colocação muito próxima do primeiro colocado.
