#**Questão1**
Crie uma função chamada filtrar_colunas_enem(df) que:
receba como entrada o DataFrame dos microdados do ENEM,
retorne apenas as colunas essenciais para análise de notas e informações básicas,
exclua as colunas referentes ao questionário socioeconômico (Q001 a Q025)

```python
#Utilizando 'drop' para excluir colunas;
#'str(i).zfill(3)' transforma cada número em string com 3 dígitos no range de 1 a 25
def filtrar_colunas_enem(df):
    return df.drop(columns=[f'Q{str(i).zfill(3)}' for i in range(1, 26)], errors='ignore')
filtrar_colunas_enem(df).head(2)
```

#**Questão 2:**
a) Crie uma tabela com as médias das notas (NU_NOTA_CN, NU_NOTA_CH, NU_NOTA_LC, NU_NOTA_MT) por estado (SG_UF_PROVA).


```python
#Filtrando colunas solicitadas
colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT']
```

```python
#Agrupamento das médias das notas por estado
tabela_medias = df.groupby('SG_UF_PROVA')[colunas_notas].mean()
tabela_medias.head(2)
```

#**Questão 2:**
 b) Identifique quais estados tiveram menor e maior média em cada área.

```python
#Selecionando menores médias e estados
menor_estado = tabela_medias.idxmin()
menor_valor = tabela_medias.min()
#Selecionando maiores médias e estados
maior_estado = tabela_medias.idxmax()
maior_valor = tabela_medias.max()
```

```python
#Apresentando menores e maiores médias por estado em dataframe
for coluna in colunas_notas:                                         #loop para cada área
    menor_estado = tabela_medias[coluna].idxmin()                    #retorna estado com menor média
    menor_valor = tabela_medias[coluna].min()                        #retorna menor média

    maior_estado = tabela_medias[coluna].idxmax()                    #retorna estado com maior média
    maior_valor = tabela_medias[coluna].max()                        #retorna maior média

    print(f"{coluna}:")                                              #imprime a área
    print(f"  Menor média → {menor_estado} ({menor_valor:.2f})")     #imprime menor média e estado 'float' 2 casas decimais
    print(f"  Maior média → {maior_estado} ({maior_valor:.2f})\n")   #imprime maior média e estado 'float' 2 casas decimais e quebra de linha
```

**Explicação** : É possível verificar que o estado do Maranhão apresentou menores médias em todas as áreas e o estado do Rio Grande do Norte apresentou maiores médias em todas as áreas. Uma apresentação visual e comparativa entre as médias dos estados é disponibilizada a seguir.

```python
#Gráfico de barras para visualizar a tabela_medias
tabela_medias.plot(kind='bar', figsize=(8, 4))                     #Escolhendo tipo de gráfico, tamanho
plt.xlabel('Estados')                                              #Nomeando eixo x
plt.ylabel('Média das Notas')                                      #Nomeando eixo y

plt.xticks(rotation=360)                                           #Rotacionando os rótulos do eixo X para evitar sobreposição
plt.legend(title='Áreas de Conhecimento', bbox_to_anchor=(1, 1))   #Adicionando uma legenda
plt.tight_layout()                                                 #Ajustando o layout para evitar que os rótulos sejam cortados
plt.show()                                                         #Chamando a imagem
```

#**Questão 3**
Calcule o desvio padrão das notas em cada área por estado e identifique, para cada estado, qual área teve maior variação. Apresente uma tabela com Estado e Área_com_maior_variação.

```python
#Calcular o desvio padrão das notas por estado
desvios = df.groupby('SG_UF_PROVA')[colunas_notas].std().round(2)
desvios
```

**Explicação** : É possível observar que a área de Matemática apresenta maiores desvios padrão para todos os estados e que os menores desvios foram apresentados para a área de Linguagens e Códigos.

```python
#Montar tabela final como dataframe apresentando área com maior variação por estado
tabela_variacao = pd.DataFrame({
    'Estado': desvios.index,
    'Area_com_maior_variacao': desvios.idxmax(axis=1).values})

tabela_variacao
```

**Explicação** : A tabela_variacao ratifica a informação que as maiores variações são referentes à área Matemática e suas Tecnologias em todos os estados estudados

#**Questão 4** : Visualização de dados
a) Crie um boxplot das notas de NU_NOTA_REDACAO por estado (Nordeste).

```python
#calculando a mediana referente à nota de redação para cada estado
medianas_redacao = df.groupby('SG_UF_PROVA')['NU_NOTA_REDACAO'].median().sort_values(ascending=False).round(2)
medianas_redacao
```

```python
#organizando dados para criação do boxplot com medianas ordenadas
red_boxplot = [df.loc[df['SG_UF_PROVA']==uf, 'NU_NOTA_REDACAO'].dropna() for uf in medianas_redacao.index]

#criação da figura boxplot
plt.figure(figsize=(12,6))
plt.boxplot(red_boxplot, labels=medianas_redacao.index)
plt.title('Redação por Estado (ordenado pela mediana)')
plt.xlabel('Estado')
plt.ylabel('Nota de Redação')
plt.show()
```

**Explicação** : O gráfico boxplot mostra que os estados do Ceará, Sergipe e Paraíba apresentaram maiores medianas quanto à área de Redação. Além disso, vale ressaltar a variação nas alturas das caixas sugerindo uma dispersão das notas (quanto maior, mais variabilidade dentro do estado). Os campeões em resíduos (outliers) foram os estados do Rio Grande do Norte e Bahia, respectivamente; onde é possível observar notas muito baixas comparadas ao padrão, revelando uma diversidade de desempenho dentro de cada estado.

#**Questão 4** : Visualização de dados
b) Crie um histograma das notas de NU_NOTA_MT para uma UF específica.

```python
#Padronização em Baixa, Média, Alta
df['MT_classe'] = np.where(df['NU_NOTA_MT'] <= 400, 'Baixa',
                   np.where(df['NU_NOTA_MT'] <= 600, 'Média', 'Alta'))

#Frequências - criando nova coluna para visualização posterior
freq = df['MT_classe'].value_counts().reindex(['Baixa','Média','Alta'])
```

```python
#Histograma
plt.figure(figsize=(7,8))
plt.bar(freq.index.astype(str), freq.values)             #Cria gráfico de barras; 'freq.index' para categorias; 'freq.values' para qtidade das frequêcias
plt.title('Histograma - Matemática e Tecnologias')
plt.ylabel('Frequência')
for i, v in enumerate(freq.values):                      #Percorre cada barra (i) e adiciona o valor (v) em cima dela
   plt.text(i, v, str(v), ha='center', va='bottom')      #Ajuste do texto e valores
plt.show()
```

**Explicação** : Observa-se que a maior frequência das notas de matemática foram classificadas entre 400 e 600 pontos.

**OBS:** Como sugestão adicional, foi realizado um ranqueamento referente às médias de Matemática entre os estados avaliados.

```python
#Ordenar a tabela de forma decrescente para cada área
tabela_medias_reset = tabela_medias.reset_index()
tabela_medias_reset = tabela_medias_reset.rename(columns={'SG_UF_PROVA': 'Estados','NU_NOTA_MT': 'Média_MT'})

#criando um ranking para as médias de Matemática
rank_mt = tabela_medias_reset[['Estados', 'Média_MT']].sort_values(by='Média_MT', ascending=False).round(2).reset_index(drop=True)
rank_mt['Posição'] = rank_mt.index + 1          #adiciona posição no ranking
rank_mt
```

```python
#Apresentação visual do ranking
plt.figure(figsize=(10,6))
plt.bar(rank_mt['Estados'], rank_mt['Média_MT'])

plt.title('Ranking das Médias de Matemática por Estado', fontsize=14)
plt.xlabel('Estados')
plt.ylabel('Média Matemática')
plt.show()
```

**Explicação** : O ranking mostra que o Estado de Rio Grande do Norte apresntou maior média para a área de Matemática e Tecnologias, seguindo do estado do Ceará com uma sugunda colocação muito próxima do primeiro colocado.
